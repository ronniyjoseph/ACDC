```python
import numpy as np
from scipy.constants import c
from matplotlib import pyplot as plt
```

```python
def otf_pupil(u, d=4, frequency=150e6):
    k = 2*np.pi*frequency/c
    response = 2/np.pi*(np.arccos(u/(k*d)) - u/(k*d)*np.sqrt(1-(u/(k*d))**2))
    response[u > k*d] = 0
    return response
```

```python
u = np.linspace(0, 20, 100)
f = np.linspace(150-15, 150+15, 100)*1e6
r = otf_pupil(u=12, d = 4, frequency=f )
plt.plot(f,r)
plt.show()
```
